```python
import numpy as np
```

```python
import pandas as pd
```

```python
np.random.seed(42) # # Random number same banane ke liye seed set kar rahe hain (taaki result baar-baar same aaye)
```

```python
apartments = [f"Apt_{i}" for i in range (1,21)] 
```

```python
dates = pd.date_range(start = '2025-07-01', periods =30 , freq = 'D')
```

```python

data = {
    "Date": np.tile(dates, len(apartments)),  # Repeat full date range for each apartment
    "Apartments": np.repeat(apartments, len(dates)),  # Repeat each apartment 30 times
    "Electricity_usage": np.random.normal(loc=70, scale=5, size=len(dates)*len(apartments))
}
data
```

```python
df = pd.DataFrame(data)
df
```

```python
df = pd.DataFrame(data)
df
```

```python
print (df.isnull().sum())
df
```

```python
print (df.describe)


print(df.dtypes)

total_usage_per_apartment = df.groupby("Apartments")["Electricity_usage"].sum()
print(total_usage_per_apartment)

usage_per_apartment = df.groupby("Apartments")["Electricity_usage"].sum().sort_values(ascending = False)
print("usage_per_apartment" , usage_per_apartment)
```

```python
daily_avg=df.groupby("Date")["Electricity_usage"].mean()
print(daily_avg.head())
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
plt.figure (figsize=(10,6))
sns.barplot(x=usage_per_apartment.index, y=usage_per_apartment.values)
plt.xticks(rotation=90)
plt.title("Total usage per Apartments")
plt.xlabel("Apartment")
plt.ylabel("Total KWH")
plt.tight_layout()
plt.show()
```

```python
#time series plot of daily average
plt.figure(figsize=(10,6))
sns.lineplot(x=daily_avg.index,y=daily_avg.values)
plt.title("Daily average electricity usage")
plt.xlabel("Data")
plt.ylabel("Average KWH")
plt.tight_layout()
plt.show()
```

```python
# Calculate threshold: mean + 2 Ã— std deviation
threshold = df["Electricity_usage"].mean() + 2 * df["Electricity_usage"].std()

# Create a new column that flags entries above the threshold
df["High_usage_flag"] = df["Electricity_usage"] > threshold

# Display the first few rows where the flag is True
print(df[df["High_usage_flag"] == True].head())
```
